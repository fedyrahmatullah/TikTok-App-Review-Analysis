```python
def tt_sentiment(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None
    
tiktok['sentiment'] = tiktok['content'].apply(tt_sentiment)
```

```python
def tt_polarity(text):
    try:
        return TextBlob(text).polarity
    except:
        return None
    
tiktok['polarity'] = tiktok['content'].apply(tt_polarity)
```

```python
def tt_subjectivity(text):
    try:
        return TextBlob(text).subjectivity
    except:
        return None
    
tiktok['subjectivity'] = tiktok['content'].apply(tt_subjectivity)
```
